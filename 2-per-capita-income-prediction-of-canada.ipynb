```python
df = pd.read_csv("../input/percapitaincomecanada/perCapitaIncome.csv")
df
```

```python
plt.scatter(df.year, df.per_capita_income_usd, marker="+")
plt.plot()
```

```python
reg = linear_model.LinearRegression()
reg.fit(df[['year']], df.per_capita_income_usd)
```

```python
reg.predict([[2022]])
```

```python
joblib.dump(reg, 'model_lib')
```

```python
model = joblib.load('model_lib')
```

```python
model.predict([[2022]])
```
